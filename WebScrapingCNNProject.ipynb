## Reading the page and Extracting All Tickers

```python
counter = 0
df = []
while counter <= 13:    # Amount of pages with Tickers
    html = browser.page_source
    data = pd.read_html(html)
    df.extend(data[0].values.tolist())  # Adding values to a new list
    
    selection = browser.find_element_by_class_name('wrsb') # Browsing a class with necessary layout with interested information
    content = selection.find_element(By.TAG_NAME, 'div') # Moving next level of tags
    nav = content.find_element(By.TAG_NAME, 'nav')  # Browsing a class with layout of buttons for clicking next page
    buttonsTicker = nav.find_elements(By.TAG_NAME, 'button') # List of 'buttons'
    counter = counter + 1
    if counter == 13:
        break
    buttonsTicker[2].click()
    
```

## The script for getting needed information about companies

```python
Tickers1 = df1.loc[df1.Indusrty == 'Banks', ['Symbol', 'Company Name']]
Tickers1
```

```python
Tickers2 = df1.loc[df1.Indusrty == 'Capital Markets', ['Symbol','Company Name']]
Tickers2
```

```python
MainLst = []
cnn_url = 'https://money.cnn.com/quote/quote.html?utm_source=quote_search&symb={}'
def Description(features):
    Result = {}
    for index, i in enumerate(features[0]): # Enumerate 0 column
        Result[i] = features[1][index]  # Assign sequential value to an appropiate value in a column. Index - where for is.
    return Result

for index, row in Tickers1.iterrows(): 
    MainDict = {}
    #print(row['Symbol'])
    company_url = cnn_url.format(row['Symbol']) # Rows with symbols for agjusting in Format Funtion
    #print(company_url)
    MainDict['Symbol'] = row['Symbol'] # Adding in the created dictionary key-value pairs
    MainDict['Company Name'] = row['Company Name']
    browser.get(company_url)
    share = browser.find_element_by_class_name('wsod_last')  # Searching share price
    costs = share.find_element(By.TAG_NAME, 'span') # Extracting the information
    #print(costs.text)
    MainDict['Share Price'] = costs.text
    
    DailyChanges = browser.find_element_by_class_name('wsod_change') # Search Daily changes
    ChangeDay = DailyChanges.find_elements(By.TAG_NAME, 'span')
    MainDict['Today\'s Change'] = ChangeDay[2].text + ChangeDay[3].text + ChangeDay[4].text # Combining three parameters
    
    AnnualChanges = browser.find_element_by_class_name('wsod_ytd') # Search Annual changes
    if len(AnnualChanges.find_elements(By.TAG_NAME, 'span')) == 0:
        pass
    else:
        ChangeYear = AnnualChanges.find_element(By.TAG_NAME, 'span')
        MainDict['Year-to-Date'] = ChangeYear.text
    
    # print(ChangeDay[2].text)
    # print(ChangeDay[3].text)
    # print(ChangeDay[4].text)
    # print(ChangeYear.text)
    
    DescriptionPage = browser.page_source     # Collecting all information
    DescriptionData = pd.read_html(DescriptionPage)
    if len(DescriptionData) == 7:
        
        # Update modifies MainDict with keys and values of DescriptionData
        
        MainDict.update(Description(DescriptionData[2]))   # Today’s Trading  
        MainDict.update(Description(DescriptionData[3]))   # Growth & Valuation
        MainDict.update(Description(DescriptionData[5]))   # Financials
    
    else:
        MainDict.update(Description(DescriptionData[1]))   # Today’s Trading
        MainDict.update(Description(DescriptionData[2]))   # Growth & Valuation
        MainDict.update(Description(DescriptionData[4]))   # Financials
     
    SelectionLink = browser.find_element_by_id('wsod_snapshotView')    # Access to articles
    SelectionPress = SelectionLink.find_element_by_id('wsod_snapshotNewsLinks')
    ButtonPRelease = SelectionPress.find_elements(By.TAG_NAME, 'a')
    ButtonPRelease[1].click() # Button for press releases
    
    Article = browser.find_element_by_class_name('wsod_dataTable') # Searching of articles
    ButtonArticle = Article.find_elements(By.TAG_NAME, 'a')  # Button for the last article
    if len(ButtonArticle) >= 1:
        Date = browser.find_element_by_id('wsod_newsAndPressReleaseContainer')    # Searching location of date
        DateExtract = Date.find_element(By.TAG_NAME, 'tr')                
        DateIssue = DateExtract.find_element(By.TAG_NAME, 'td').text      # Extraction of date
        
        ButtonArticle[0].click() # Button for choosing an article
        
        Waiting = ''
        while not Waiting:   # While loop to make sure that page with article is loaded and scripts does not go further
            try:
                Waiting = browser.find_element_by_class_name('content_container') # This class exists only on a page with articles
    
            except:
                continue
    
        ArticleInfo = BeautifulSoup(browser.page_source, 'html.parser')
        Text = ArticleInfo.find('div', {'class':'column' }) # Browsing a class with necessary layout of article
        TitleArt = Text.find('h1').text # Extracting information
        Content = Text.text.strip() 
        
        MainDict['Release Date'] = DateIssue
        MainDict['Title'] = TitleArt
        MainDict['Article'] = Content
        
    print(MainDict)
    MainLst.append(MainDict)
```
